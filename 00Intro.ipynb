```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def y(x):
    return 8*(0.5 + np.sin(x)/x)
x = np.linspace(2,9.5)
```

```python
fig, ax = plt.subplots()
ax.plot(x, y(x), 'b')
ax.set_xlim(0,10)
ax.set_ylim(0,10)
ax.set_xticks([])
ax.set_yticks([])
ax.set_aspect('equal')
ax.plot([x[0]], [y(x[0])], 'bo')
ax.plot([x[-1]], [y(x[-1])], 'bo')
ax.text(x[0], 1.05*y(x[0]), 'A', fontsize=15)
ax.text(0.95*x[-1], 0.8*y(x[-1]), 'B', fontsize=15);
```

```python
fig, ax = plt.subplots()
ax.plot(x, y(x), 'b')
ax.set_xlim(0,10)
ax.set_ylim(0,10)
ax.grid(True)
ax.set_aspect('equal')
ax.set_xlabel('x [cm]')
ax.set_ylabel('y [cm]')
ax.plot([x[0]], [y(x[0])], 'bo')
ax.plot([x[-1]], [y(x[-1])], 'bo')
ax.text(x[0], 1.05*y(x[0]), 'A ({:.1f},{:.1f})'.format(x[0],y(x[0])), fontsize=15)
ax.text(0.95*x[-1], 0.8*y(x[-1]), 'B', fontsize=15)
ax.text(0.75*x[-1], 0.6*y(x[-1]), '({:.1f},{:.1f})'.format(x[-1],y(x[-1])), fontsize=15);
```

```python
fig, ax = plt.subplots()
ax.plot(x, y(x), 'b')
ax.set_xlim(0,10)
ax.set_ylim(0,10)
ax.grid(True)
ax.set_aspect('equal')
ax.set_xlabel('x [cm]')
ax.set_ylabel('y [cm]')
ax.plot([x[0]], [y(x[0])], 'bo')
ax.plot([x[-1]], [y(x[-1])], 'bo')
#ax.text(x[0], 1.05*y(x[0]), 'A', fontsize=15)
#ax.text(0.95*x[-1], 0.8*y(x[-1]), 'B', fontsize=15);
soa = np.array([[0,0,x[0],y(x[0])], [0,0,x[-1],y(x[-1])]])
X,Y,U,V = zip(*soa)
ax.quiver(X,Y,U,V,angles='xy',scale_units='xy',scale=1)
ax.text(0.5, 5.2, r'$\vec{r}_i$', fontsize=15)
ax.text(5, 1, r'$\vec{r}_f$', fontsize=15);
```

```python
fig, ax = plt.subplots()
ax.plot(x, y(x), 'b')
ax.set_xlim(0,10)
ax.set_ylim(0,10)
ax.grid(True)
ax.set_aspect('equal')
ax.set_xlabel('x [cm]')
ax.set_ylabel('y [cm]')
ax.plot([x[0]], [y(x[0])], 'bo')
ax.plot([x[-1]], [y(x[-1])], 'bo')
#ax.text(x[0], 1.05*y(x[0]), 'A', fontsize=15)
#ax.text(0.95*x[-1], 0.8*y(x[-1]), 'B', fontsize=15);
soa = np.array([[0,0,x[0],y(x[0])], [0,0,x[-1],y(x[-1])], [x[0],y(x[0]),x[-1]-x[0],y(x[-1])-y(x[0])]])
X,Y,U,V = zip(*soa)
ax.quiver(X,Y,U,V,angles='xy',scale_units='xy',scale=1)
ax.text(0.5, 5.2, r'$\vec{r}_i$', fontsize=15)
ax.text(5, 1, r'$\vec{r}_f$', fontsize=15);
ax.text(5, 6.5, r'$\vec{r}_f - \vec{r}_i$', fontsize=15);
```

```python
def y2(x):
    return y(x)-5
x2 = x-5
```

```python
fig, ax = plt.subplots()
ax.plot(x2, y2(x), 'b')
ax.set_xlim(-5,5)
ax.set_ylim(-5,5)
ax.grid(True)
ax.set_aspect('equal')
ax.set_xlabel('x [cm]')
ax.set_ylabel('y [cm]')
ax.plot([x2[0]], [y2(x[0])], 'bo')
ax.plot([x2[-1]], [y2(x[-1])], 'bo')
ax.text(x2[0], 1.1*y2(x[0]), 'A ({:.1f},{:.1f})'.format(x2[0],y2(x[0])), fontsize=15)
ax.text(0.95*x2[-1], 1.8*y2(x[-1]), 'B', fontsize=15)
ax.text(0.4*x2[-1], 2.5*y2(x[-1]), '({:.1f},{:.1f})'.format(x2[-1],y2(x[-1])), fontsize=15);
```
